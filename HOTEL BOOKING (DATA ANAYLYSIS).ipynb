```python
#NOW WE WILL FIND THAT WHAT IS THE PERCENTAGE OF CANCELLATION .
#VALUE.COUNT() FUNCTION HELPS TO GET OBJECT NAME OF PARTICULAR COLOUMN AND AMOUNT OF TIME THAT OBJCT IS PRESENT
#NORMALIZE=TRUE IS USED TO FIND THE PERCENTAGE|
```

```python
cancelled_perc=df['is_canceled'].value_counts(normalize='true')
cancelled_perc
```

```python
print(cancelled_perc)
plt.figure(figsize=(5,5)) # size of box
plt.title('Reservation status count')
plt.bar(['not cancelled' , 'cancelled'] ,df['is_canceled'].value_counts(), edgecolor ='k',width=0.7)
plt.show()
```

```python
#now depending upon the hotels we have to find whose cancellation rate is more and whose less
```

```python
#seaborn we will use for data visualization 
#seaborn gives us benefit in It provides a high-level interface for drawing attractive and informative statistical graphics.
#hue means on which basis we count 
#legend labels means jo side mein box rehta chota sa usme likha hota ki kahe ka graph hai
```

```python
plt.figure(figsize=(8,4))
ax1= sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in diffrent hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
#if we want to focus on resort hotel like how many % of resort hotel is cacnelled 
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']# we filterd only resort hotel in hotel coloumn
resort_hotel['is_canceled'].value_counts(normalize= True)
```

```python
#now to check on city hotels
```

```python
city_hotel=df[df['hotel']=='City Hotel']# we filterd only resort hotel in hotel coloumn
city_hotel['is_canceled'].value_counts(normalize= True)
```

```python
#we can see cancellattion rate is too much high on city hotel as compared to resort hotel.
#we will check whether price is the issue of cancellation
#group by function is used because ek din mein kaafi booking hoti to ek din k average daily rate ka mean lenge 
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#now we create visualisation of average daily rate
#index mean reservation day
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate inCity and Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'Resort hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label = 'city hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
    #now we check on month basis is reservation status per momth
    #humare pass koi month wala coloumn nhi hai to we will exctract month from reservatiion status date then we will plot.
```

```python
df['month'] =df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month' , hue='is_canceled',data=df,palette='bright')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
#we can see august ie 8th month mein more reservation and less cancellation..aaisa to nhi ki price kam ho tab
#soo check krna padega harr month pe price 
# we can do this by the help of average dily rate adr
#visualtion jo krne wale vo sirf cancelation data pe  check kreinge bcoz hme ye check krna ki price ka cancellation date pe koi effect hai ki nhi
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())

plt.show()
```

```python
#adr is lowest in august and also cancellation rate is low in august
#adr is high in january and also cancellation rate is high in january
#its prove our hypothesis ki price affect the cancellation rate 
```
